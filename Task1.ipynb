ðŸ”¹âœ… Step 1: Load the Dataset

```python
import pandas as pd

# Load the CSV with skipped metadata rows
df = pd.read_csv(
    "C:/Users/Aishwarya/Downloads/API_SP.POP.TOTL_DS2_en_csv_v2_26346/API_SP.POP.TOTL_DS2_en_csv_v2_26346.csv",
    skiprows=4
)

# Preview
print(df.head())
```

ðŸ”¹âœ… Step 2: Clean the Data

```python
df = df.drop(columns=['Unnamed: 68'], errors='ignore')
```

ðŸ”¹âœ… Step 3: Choose What to Visualize

âž¤ Extract 2022 data

```python
df_2022 = df[['Country Name', '2022']].dropna()
```

âž¤ Get top 10 countries by population

```python
top10 = df_2022.sort_values(by='2022', ascending=False).head(10)
```

ðŸ”¹âœ… Step 4: Plot the Bar Chart

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(top10['Country Name'], top10['2022'], color='skyblue')
plt.xticks(rotation=45)
plt.xlabel('Country')
plt.ylabel('Population in 2022')
plt.title('Top 10 Most Populated Countries (2022)')
plt.tight_layout()
plt.show()
```

```python
ðŸ”¹ âœ… Step 5: Histogram Instead
```

```python
plt.figure(figsize=(10,6))
plt.hist(df_2022['2022'], bins=20, color='lightcoral', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.title('Distribution of Population Across Countries (2022)')
plt.tight_layout()
plt.show()
```
